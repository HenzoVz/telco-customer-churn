### About our Dataset
Source - https://www.kaggle.com/blastchar/telco-customer-churn
1. Temos informações de clientes para uma empresa de telecomunicações
2. Temos IDs de clientes, informações gerais sobre clientes, serviços que eles também assinaram, tipo de contrato e cobranças mensais.
3. Como são informações históricas do cliente, temos um campo informando se esse cliente possui **churnded** 
**Field Descriptions**
- customerID - Identificação do Cliente
- gender - Se o cliente é homem ou mulher
- SeniorCitizen - Se o cliente é um idoso ou não (1, 0)
- Partner - Se o cliente tem ou não um parceiro (Sim, Não)
- Dependents - Se o cliente tem dependentes ou não (Sim, Não)
- tenure - Número de meses em que o cliente ficou na empresa
- PhoneService - Se o cliente tem um serviço telefônico ou não (Sim, Não)
- MultipleLines - Se o cliente possui várias linhas ou não (Sim, Não, Sem serviço telefônico)
- InternetService - Fornecedor de serviços de Internet do cliente (DSL, Fiber optic, No)
- OnlineSecurity - Se o cliente possui segurança on-line ou não (Yes, No, No internet service)
- OnlineBackup - Se o cliente possui backup on-line ou não (Yes, No, No internet service)
- DeviceProtection - Se o cliente possui proteção para o dispositivo ou não (Yes, No, No internet service)
- TechSupport - Se o cliente tem suporte técnico ou não (Yes, No, No internet service)
- StreamingTV - Se o cliente tem ou não streaming de TV (Yes, No, No internet service)
- StreamingMovies - Se o cliente tem streaming de filmes ou não (Yes, No, No internet service)
- Contract - O prazo do contrato do cliente (Month-to-month, One year, Two year)
- PaperlessBilling - Se o cliente possui faturamento sem papel ou não (Yes, No)
- PaymentMethod - O método de pagamento do cliente (Electronic check, Mailed check Bank transfer (automatic), Credit card (automatic))
- MonthlyCharges - O valor cobrado mensalmente ao cliente
- TotalCharges - O valor total cobrado ao cliente
- Churn - Se o cliente agitou ou não (Yes or No)

***Customer Churn*** - rotatividade é quando um cliente, usuário, jogador, assinante existente ou qualquer outro tipo de cliente de retorno deixa de fazer negócios ou encerra o relacionamento com uma empresa..
